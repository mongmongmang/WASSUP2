```python
#coffee.py
coffee = 10
while True:
    money = int(input("돈을 넣어 주세요: "))
    if money == 300:
        print("커피를 줍니다.")
        coffee = coffee -1
        print("남은 커피의 양은 %d개 입니다." % coffee)
    elif money > 300:
        print("거스름돈 %d를 주고 커피를 줍니다." % (money -300))
        coffee = coffee -1
        print("남은 커피의 양은 %d개 입니다." % coffee)
    else:
        print("돈을 다시 돌려주고 커피를 주지 않습니다.")
        print("남은 커피의 양은 %d개 입니다." % coffee)
    if coffee == 0:
        print("커피가 다 떨어졌습니다. 판매를 중지 합니다.")
        break
```

## while 문의 맨 처음으로 돌아가기  
- continue : while 문을 빠져나가지 않고 while 문의 맨 처음(조건문)으로 다시 돌아가게 만들고 싶은 경우
- while은 조건에 맞지 않으면 빠져나옴

```python
a = 0
while a < 10:
    a = a + 1 #a +=1
    if a % 2 == 0: continue #짝수일 경우 print하지 않고 다시 돌아감
    print(a)  #홀수일때, print
```

for와 range 함수를 사용하면 소스 코드 단 4줄만으로 구구단을 출력할 수 있다. 들여쓰기에 주의하면서 입력해 보자.

```python
for i in range(2,10):  #i가 선택된 상태에서 j실행 -반복
        print(i)
        for j in range(1,10):
            print(j)
```

## 리스트 컴프리헨션 (표현식)
- 그냥 알면 좋고 몰라도됨
- 속도가 2배차이, 퍼포먼스 좋음
- 정말 많은 연산, 인공지능 등 할 때 사용

```python
>>> a = [1,2,3,4]
>>> result = [num * 3 for num in a] # for num in a 가 num*3
>>> print(result)
[3, 6, 9, 12]
```

```python
a = [1,2,3,4]
>>> result = [num * 3 for num in a if num % 2 == 0]  #if 조건이 해당되면, num*3을 실행시킨다
>>> print(result)
```

### 시간 측정을 위한 코드, 2배넘게 차이남

```python

import time

start = time.time() 
numbers = [i for i in range(1000000)]  #컨프리헨션
end = time.time() 
print('List Comprehension Time:', end-start)

start = time.time() 
numbers = []
for i in range(1000000):   #for loop
    numbers.append(i)
end = time.time() 
print('For loop Time:', end-start)
```
