- 계산기는 이전에 계산한 결과값을 기억하고 있어야한다.

```python
# calculator.py
result = 0

def add(num):
    global result # 전에 계산한 결괏값을 유지하기 위해서 result 전역 변수(global)를 사용
    result += num  # 결괏값(result)에 입력값(num) 더하기 #전협변수에 메모리가 유지됨
    return result  # 결괏값 리턴

print(add(3))
print(add(4))
```

```python
#만약 한프로그램에서 2대의 계산기가 필요한 상황이라면?
# calculator2.py
result1 = 0
result2 = 0

def add1(num):  # 계산기1
    global result1
    result1 += num
    return result1

def add2(num):  # 계산기2
    global result2
    result2 += num
    return result2

print(add1(3))
print(add1(4))
print(add2(3))
print(add2(7))
# 서로 독립적이기 위해 함수를 따로 만든 것임
```

클래스 쓰는 이유 :

```python
# calculator3.py
class Calculator:
    def __init__(self):
        self.result = 0

    def add(self, num):
        self.result += num
        return self.result

cal1 = Calculator() #cal1이라는 인스턴스를 만들어냄
cal2 = Calculator()

print(cal1.add(3))
print(cal1.add(4))
print(cal2.add(3))
print(cal2.add(7))
```

## try-except 문

### 1. try-except만 쓰는 방법 

```python
try:                   
    ...                             # 모니터링 블록 (오류가능성 0 -> except 실행)  
except [발생오류 [as 오류변수]]:   # except 블록(이 기호는 괄호 안의 내용을 생략할 수 있다는 관례적인 표기법)
    ...                             # 오류가 없다면 except 다음 블록 실행
```

```python
# 1. try-except만 쓰는 방법 
#   : 다 잡는다! 
try:
    ...
except:
    ...  
```

```python
# 2. 발생 오류만 포함한 except 문 (분기만 0)   Q. 분기가 무슨 뜻이죠?
#    : except 문에 미리 정해 놓은 오류인 경우에만 except 블록을 수행
try:
    ...
except 발생오류:
    ...
```

```python
# 3. 발생 오류와 오류 변수까지 포함한 except 문 [가장많이쓰임**]
#    : 2. 의 경우에서 + 발생오류를 변수로 받아올 수 있음 (분기+변수로)
try:
    ...
except 발생오류 as 오류변수:    
    ...
```

### 예시

```python
# try_except.py
try:
    4 / 0
except :
    print('error')   # 만약에 에러가 나면, 에러말고 프린트를 실행시켜줘
                     # -> (정상동작) 결론적으로 오류가 아닌 실행문이 됨
```

```python
# try_except.py
try:
    4/0
except ZeroDivisionError as e:    #지정한 해당 '에러정보'만 분기하여 변수로 받아올 수 있음
    print(e)                       #에러정보 출력
```

### 에러로 정의 하느냐 / 안하느냐(중요)
팝업 등으로 에러를 안내하기 위해 3번이 사용됨***  
실제 에러 - 프리징, 꺼지기 / 에러지만 정상작동 - 팝업, 다이어로그, 스낵바, 투스트 등  
- 회원가입 유효성 검사, 작은이슈들은 에러로 정의하지 않음  
    ex. 분기처리해서 고객에게 안내 (팝업, 토스트, 스낵바 등)  
    ex. [UIUX] B2C :투스트, 스낵바와 같이 안내 / B2B,은행 등 : 팝업   

## 여러 개의 오류 처리하기

```python
try:
    ...
except 발생오류1:
   ... 
except 발생오류2:
   ...
```

```python
# many_error.py : 여러 개의 오류처리시  (에러는 자세하게 적는 게 좋음)
try:
    a = [1, 2]
    print(a[3])           
    4/0
except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.")
except IndexError:            # Q.파이썬은 순서대로, except print하면 빠져나오는 건가요? 
    print("인덱싱 할 수 없습니다.")
```

```python
try:
    a = [1,2]
    print(a[3])
    4/0
except ZeroDivisionError as e:
    print(e)
except IndexError as e:
    print(e)
```

```python
# 두개 이상가능 간략한 코드 (잘안씀)
try:
    a = [1,2]
    print(a[3])
    4/0
except (ZeroDivisionError, IndexError) as e:
    print(e)
```

## Try-else문
명확히 하려면 else 작성추천, else 안써도 작동 0

```python
try:
    ...
except [발생오류 [as 오류변수]]:
    ...
else:  # 오류가 없을 경우에만 수행
    ...
```

```python
# try_else.py 
try :
    age = int(input('나이를 입력하세요 : '))
except : 
    print('입력이 정확하지 않습니다.')
   # int(input('나이를 입력하세요 : '))  # Q. 입력이 정확하지 않습니다-> 입력창 나오게 하려면
else :
    if age <= 18:
        print('미성년자는 출입금지입니다.')
    else:
        print('환영합니다.')
```
