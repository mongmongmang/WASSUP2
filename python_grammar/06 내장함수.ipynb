## all
동일한 형식인지 참/거짓

반복 가능한 데이터 x를 입력값으로 받으며 이 x의 요소가 모두 참이면 True, 거짓이 하나라도 있으면 False를 리턴한다.

반복 가능한 데이터란 for 문에서 사용할 수 있는 자료형을 의미한다. 리스트, 튜플, 문자열, 딕셔너리, 집합 등이 있다.

```python
#all 
>>> all([1, 2, 3])
True
```

```python
>>> all([1, 2, 3, 0])
False
# 0은 거짓
```

```python
>>> all([])
True
```

## filter**(중요)
- 데이터 전처리시 많이 사용됨

```python
# positive.py 
def positive(l): 
    result = [] 
    for i in l: 
        if i > 0: 
            result.append(i) 
    return result

print(positive([1,-3,2,0,-5,6]))
```

```python
# filter1.py
def positive(x):
    return x > 0

print(list(filter(positive, [1, -3, 2, 0, -5, 6])))
```

```python
filter(positive, [1, -3, 2, 0, -5, 6]) 
```

```python
list(filter(positive, [1, -3, 2, 0, -5, 6]))
```

```python
>>> list(filter(lambda x: x > 0, [1, -3, 2, 0, -5, 6]))
[1, 2, 6]
```
